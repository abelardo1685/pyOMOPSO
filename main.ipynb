## Visualization part

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from Mat_cost import *
```

```python
x = []
y = []
z = []

costos = mat_cost(e_file)

for i in range(0,len(e_file)):

    mat_vector = costos[:,0]
    cost_vector = mat_vector[i]
    x.append(cost_vector[0])
    y.append(cost_vector[1])
    z.append(cost_vector[2])
    
# Creating figure
fig = plt.figure(figsize = (10, 7))
ax = plt.axes(projection ="3d")
 
# Creating plot
ax.scatter3D(x, y, z, color = "green")
plt.title("simple 3D scatter plot")
 
# show plot
plt.show()
```
